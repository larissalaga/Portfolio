```python
!pip install --quiet -r requirements.txt

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Lendo o arquivo CSV

```python
producao_plastico = pd.read_csv('../1- producao-de-plastico-global.csv')
producao_plastico
```

## 3. Removendo dados faltantes do arquivo CSV e salvando na variável "dpp_clean"

```python
dpp_clean = producao_plastico.dropna()
dpp_clean.info()
```

```python

dpp_clean.to_csv('../dataframes/1-producao-plastico.csv', index=False)
```

## 4. Análise de um Gráfico

Ao observamos uma amostra dos dados acima, pudemos ver uma co-relação entre o ano e a quatidade de plástico produzida


```python
fig, ax = plt.subplots(figsize=(100,80))
sns.barplot(data=dpp_clean, x="Ano", y="Produção Anual de Plástico", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=60)
plt.tight_layout()
plt.show()
```

Através da análise do gráfico, pudemos observar que conforme os anos foram passando aumentou-se a produção de plástico.
