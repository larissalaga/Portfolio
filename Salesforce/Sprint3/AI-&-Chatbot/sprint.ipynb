```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
df = pd.read_csv("salesforce.csv")
```

```python
df.sample(5)
```

```python
sns.scatterplot(
    df,
    x="X Position",
    y="Y Position",
    c="k"
)
```

```python
imagem = cv2.imread("site.jpeg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem)

df_resolucao = df.copy()

df_resolucao["X Position"] = df_resolucao["X Position"] * 0.85
df_resolucao["Y Position"] = df_resolucao["Y Position"] * 0.85

sns.scatterplot(
    df_resolucao,
    x="X Position",
    y="Y Position",
    c="k",
    s=2
)

sns.kdeplot(
    df_resolucao,
    x="X Position",
    y="Y Position",
    fill=True,
    alpha=.5,
    cmap="rocket"
)

plt.xlim(0, imagem.shape[1])
plt.ylim(imagem.shape[0], 0)

plt.axis("off")
plt.savefig("pontos.png", dpi=1000)
plt.show
```

### Conclusões

É possível considerar que os pontos mais quentes do site são o carrossel, a seção com leve explicação sobre os produtos e o cabeçalho para fazer a transição entre a página principal e a inicial.
