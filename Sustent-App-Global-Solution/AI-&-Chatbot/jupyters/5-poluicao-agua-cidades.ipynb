```python
!pip install --quiet -r requirements.txt

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Lendo o arquivo CSV

```python
poluicao_agua = pd.read_csv('../5- poluicao-agua-cidades.csv')
poluicao_agua
```

## 3. Removendo dados faltantes do arquivo CSV e salvando na variável "pac_clean"

```python
pac_clean = poluicao_agua.dropna()
pac_clean.info()
```

```python
pac_clean.to_csv('../dataframes/5-poluicao-agua.csv', index=False)
```

## 4. Análise de um Gráfico

Ao observamos uma amostra dos dados acima, pudemos ver uma co-relação entre a entidade e a poluição da água 

```python
fig, ax = plt.subplots(figsize=(200,70))
sns.barplot(data=pac_clean, x=" Entidade", y=" Poluição da Água", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=60)
plt.tight_layout()
plt.show()
```

Através da análise do gráfico, pudemos observar um recorte de como é a água em alguns lugares
