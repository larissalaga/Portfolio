## 2. Lendo o arquivo CSV

```python
destino_plastico = pd.read_csv('../3- destino-plastico.csv')
destino_plastico
```

## 3. Removendo dados faltantes do arquivo CSV e salvando na variável "dp_clean"

```python
dp_clean = destino_plastico.dropna()
dp_clean.info()
```

```python
dp_clean.to_csv('../dataframes/3-destino-plastico.csv', index=False)
```

```python
print(dp_clean.columns)
```

## 4. Análise de um Gráfico

Ao observamos uma amostra dos dados acima, pudemos ver uma co-relação entre o ano, as entidades e a participação delas na reciclagem do lixo


```python
fig, ax = plt.subplots(figsize=(100,80))
sns.barplot(data=dp_clean, x=" Ano", y=" Participação da reciclagem do lixo total regional", hue="Entidade", ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=60)
plt.tight_layout()
plt.show()
```

Através da analise do gráfico pudemos ver como foi a participação de alguns países ou do mundo no geral na reciclagem do lixo
