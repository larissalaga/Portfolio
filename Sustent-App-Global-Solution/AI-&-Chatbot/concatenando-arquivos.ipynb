```python
!pip install --quiet -r requirements.txt

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df1 = pd.read_csv("dataframes/1-producao-plastico.csv")
df2 = pd.read_csv("dataframes/2-participacao-despejo-residuo.csv")
df3 = pd.read_csv("dataframes/3-destino-plastico.csv")
df4 = pd.read_csv("dataframes/4-desperdicio-per-capita.csv")
df5 = pd.read_csv("dataframes/5-poluicao-agua.csv")
```

```python
merged_df = pd.merge(df1, df2, on=['Entidade','Ano'], how='outer')
merged_df = pd.merge(merged_df, df3, on=['Entidade','Ano'], how='outer')
merged_df = pd.merge(merged_df, df4, on=['Entidade','Ano'], how='outer')
merged_df = pd.merge(merged_df, df5, on='Entidade', how='outer')
merged_df.to_csv("./dataframes/resultado-final-dataframes.csv", index=False)

```
